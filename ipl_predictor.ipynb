```python
total_score_df = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df['total_runs'] = total_score_df['total_runs']+1
```

```python
total_score_df = total_score_df[total_score_df['inning']==1]
```

```python
total_score_df
```

```python
match_df = matches.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
## many teams are replaced of removed since 2009 so we will change the dataset accordingly
```

```python
# 1) replacing deccan chargers with sunrisers and replacing delhi darevils with delhi capitals 
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

deliveries['batting_team'] = deliveries['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
deliveries['bowling_team'] = deliveries['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df['team1'].unique()
```

```python
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

deliveries['batting_team'] = deliveries['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
deliveries['bowling_team'] = deliveries['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df['team2'].unique()
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians',
        'Royal Challengers Bangalore','Kolkata Knight Riders',
        'Delhi Capitals', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals', ]
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.team1.unique()
```

```python
# only matches which have dl_review = 0 bascially only matches which were not interrupted by rain
```

```python
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df.shape
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
deliveries_df = match_df.merge(deliveries,on='match_id')
```

```python
deliveries_df.head()
```

```python
deliveries_df.batting_team.unique()
```

```python
deliveries_df = deliveries_df[deliveries_df['inning']==2]
```

```python
deliveries_df.head()
```

```python
deliveries_df['current_score'] = deliveries_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
deliveries_df
```

```python
deliveries_df.rename(columns={'total_runs_x':'target'},inplace=True)
```

```python
deliveries_df['runs_left'] = deliveries_df['target']-deliveries_df['current_score']
```

```python
deliveries_df
```

```python
deliveries_df['balls remaining'] = 126 - deliveries_df['over']*6 - deliveries['ball']
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna(0)
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].apply(lambda x:x if x==0 else 1)
deliveries_df['wickets left'] = 10 - deliveries_df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
deliveries_df['current run rate'] = deliveries_df['current_score']*6/(120-deliveries_df['balls remaining'])
```

```python
deliveries_df['required run rate'] = deliveries_df['runs_left']*6/deliveries_df['balls remaining']
```

```python
deliveries_df = deliveries_df[deliveries_df['match_id']!= 11415]
```

```python
deliveries_df['result'] = deliveries_df['winner']==deliveries_df['batting_team']
```

```python
deliveries_df['result'] = deliveries_df['result'].replace(True,1)
deliveries_df['result'] = deliveries_df['result'].replace(False,0)
```

```python
deliveries_df = deliveries_df[['batting_team','bowling_team','city','runs_left','balls remaining','wickets left','target','current run rate','required run rate','result']]
```

```python
final_df = deliveries_df
```

```python
final_df
```

```python
final_df.dropna(inplace = True)
```

```python
final_df = final_df[final_df['balls remaining']!=0]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
y = final_df['result']
X = final_df.drop(columns='result',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop = 'first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
trf
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
